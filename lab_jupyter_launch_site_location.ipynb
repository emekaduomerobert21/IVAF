*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red

# Define a function to assign marker colors based on the `class` value
def assign_marker_color(launch_class):
    if launch_class == 1:
        return 'green'
    else:
        return 'red'

# Apply the function to create the `marker_color` column
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)

# Display the first few rows to verify the new column
spacex_df.head()
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
from folium.plugins import MarkerCluster

# Initialize MarkerCluster
marker_cluster = MarkerCluster()

# Add marker_cluster to the current site_map
site_map.add_child(marker_cluster)

# Iterate over each row in spacex_df
for index, record in spacex_df.iterrows():
    # Extract the coordinates and marker color
    coordinate = [record['Lat'], record['Long']]
    marker_color = record['marker_color']
    
    # Create a Marker object with customized icon property
    marker = folium.Marker(
        location=coordinate,
        icon=folium.Icon(color='white', icon_color=marker_color),
        popup=f"Launch Site: {record['Launch Site']}<br>Outcome: {'Success' if record['class'] == 1 else 'Failure'}"
    )
    
    # Add the marker to the marker cluster
    marker_cluster.add_child(marker)

# Display the map
site_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
# Add a column for each point of interest's distance


# Import the necessary libraries
from geopy.distance import geodesic
import pandas as pd

# Define coordinates for points of interest
# For example: NASA Johnson Space Center and other notable locations
points_of_interest = {
    'NASA Johnson Space Center': (29.559684888503615, -95.0830971930759),  # NASA JSC coordinates
    'Cape Canaveral': (28.562302, -80.577356),  # CCAFS LC-40 coordinates
    'Kennedy Space Center': (28.573255, -80.646895),  # KSC LC-39A coordinates
}

# Sample DataFrame: launch_sites_df (Assume it is already created)
launch_sites_df = pd.DataFrame({
    'Launch Site': ['CCAFS LC-40', 'CCAFS SLC-40', 'KSC LC-39A', 'VAFB SLC-4E'],
    'Lat': [28.562302, 28.563197, 28.573255, 34.632834],
    'Long': [-80.577356, -80.576820, -80.646895, -120.610746],
})

# Iterate over each point of interest and calculate the distance to each launch site
for place, coord in points_of_interest.items():
    # Calculate the distance and add it as a new column to the DataFrame
    launch_sites_df[f'Distance to {place} (km)'] = launch_sites_df.apply(
        lambda row: geodesic((row['Lat'], row['Long']), coord).km, axis=1
    )

# Display the updated DataFrame with distances
launch_sites_df
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

# Analysis of Launch Sites and Their Proximities
## After plotting the distance lines from launch sites to various proximities (railways, highways, coastlines, and cities), we can analyze the patterns:

1. Are launch sites in close proximity to railways?
Observation:
Most launch sites are located near railways. This proximity is likely due to the need to transport large and heavy rocket components and fuel, which are better handled via rail.
Example:
Kennedy Space Center has a railway line approximately 12 km away.
2. Are launch sites in close proximity to highways?
Observation:
Launch sites are usually near highways to facilitate personnel movement, equipment delivery, and emergency access.
Example:
Cape Canaveral is situated close to several highways, which ensures logistical support.
3. Are launch sites in close proximity to coastlines?
Observation:
Launch sites are often located near coastlines to allow for safer rocket trajectories over open water, minimizing risks to populated areas in case of launch failures.
Example:
All SpaceX launch sites in the dataset (e.g., Cape Canaveral, Kennedy Space Center) are within a few kilometers of the coastline.
4. Do launch sites keep certain distances away from cities?
Observation:
Yes, launch sites are generally located far from cities to mitigate risks associated with launch failures, debris, or explosions.
Example:
Cape Canaveral is over 10 km away from significant urban areas.

# Findings
## Safety and Logistics:

1. Proximity to coastlines and distance from cities align with safety considerations.
2. Proximity to railways and highways supports logistical operations.
Geographic Constraints:

Sites near the equator (e.g., Cape Canaveral) provide additional benefits for launches to geostationary orbits, leveraging the Earth's rotational speed.
Optimized Placement:

The launch sites' locations reflect a balance between safety, logistics, and efficiency.
